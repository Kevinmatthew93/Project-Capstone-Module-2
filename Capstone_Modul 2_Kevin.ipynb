### 7.4  Mengisi Missing Values pada `payAmount`
Menangani Missing Values pada kolom `payAmount` berdasarkan kolom `corridorID`. Terdapat tiga jenis moda transportasi pada `corridorID` dengan nilai `payAmount` yang berbeda setiap jenisnya,
- Moda TransJakarta                : 3.500
- Moda Royaltrans                  : 20.000                
- Moda Mikrotrans                  : 0 

Berdasarkan hasil pencarian pada website TransJakarta.co.id didapatkan informasi karakteristik `corridorID` sebagai berikut :
- Moda TransJakarta                 : Diawali dengan angka kemudian diikuti dengan huruf alfabet 
- Moda Royaltrans                   : '1K', 'B13', 'D32', 'S14', '1T', 'B14', 'S12', 'S31', '6P', 'D31', 'S13', 'T21'
- Moda Mikrotrans                   : Diawali dengan kode 'JAK'

```python
# Fungsi untuk mengisi kolom payAmount
def fill_pay_amount(corridorID):
    if corridorID.startswith('JAK'):
        return 0
    elif corridorID in ['1K', 'B13', 'D32', 'S14', '1T', 'B14', 'S12', 'S31', '6P', 'D31', 'S13', 'T21']:
        return 20000
    else:
        return 3500

# Menambahkan kolom payAmount ke DataFrame
df['payAmount'] = df['corridorID'].apply(fill_pay_amount)
```

### 8.1 Analisis Tren Jumlah Penumpang Mingguan April 2023  
Menghitung rata-rata jumlah penumpang setiap hari dari hari Senin sampai Minggu pada bulan April 2023. Rata-rata jumlah penumpang dihitung dari variabel `tapInTime`. Kemudian dilakukan visualisasi menggunakan line chart.

```python
# Menambahkan kolom hari
df['day'] = df['tapInTime'].dt.day_name()

# Menentukan urutan hari dari Monday ke Sunday
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df['day'] = pd.Categorical(df['day'], categories=days_order, ordered=True)

# Menghitung jumlah penumpang setiap hari
daily_passenger_count = df.groupby('day')['tapInTime'].count().reset_index()
daily_passenger_count.columns = ['Hari', 'Jumlah Penumpang']

# Mengurutkan berdasarkan urutan hari
daily_passenger_count = daily_passenger_count.sort_values('Hari').reset_index(drop=True)

daily_passenger_count
```

```python
# Ekstrak hari dari tapInTime
df['day'] = df['tapInTime'].dt.day_name()  # Mengambil nama hari

# Hitung jumlah penumpang per hari
passenger_count = df.groupby('day')['tapInTime'].count().reset_index()

# Urutkan berdasarkan hari
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
passenger_count['day'] = pd.Categorical(passenger_count['day'], categories=day_order, ordered=True)
passenger_count = passenger_count.sort_values('day')

# Menambahkan kolom kategori warna
passenger_count['color'] = passenger_count['day'].apply(lambda x: 'weekday' if x in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'] else 'weekend')

# Visualisasikan menggunakan bar chart dengan Seaborn
plt.figure(figsize=(10, 6))
palette = {'weekday': '#FFB300', 'weekend': '#D50032'}
sns.barplot(data=passenger_count, x='day', y='tapInTime', hue='color', palette=palette)
plt.title('Tren Jumlah Penumpang Mingguan April 2023')
plt.xlabel('Hari')
plt.ylabel('Jumlah Penumpang')
plt.xticks(fontsize=10)
plt.legend()
plt.show()
```

### 8.2 Analisis Tren Jumlah Penumpang per jam pada Hari Kerja
Berdasarkan hasil analisa Mingguan, dilakukan analisa lebih lanjut untuk mengetahui tren rata-rata jumlah penumpang per jam di hari kerja, yakni hari Senin sampai Jumat.

```python
def jumlah_penumpang_per_jam_weekday(df):
    # Ekstrak hari dari 'tapInTime'
    df['day'] = df['tapInTime'].dt.day_name()  # Mengambil nama hari
    df['hour'] = df['tapInTime'].dt.hour  # Mengambil jam dari tapInTime

    # Filter hanya weekday (Senin sampai Jumat)
    weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
    df_weekday = df[df['day'].isin(weekday)]

    # Hitung jumlah penumpang per jam pada weekday
    passenger_count_hour = df_weekday.groupby(['hour'])['tapInTime'].count().reset_index()
    
    # Ubah nama kolom
    passenger_count_hour.rename(columns={'hour': 'Jam', 'tapInTime': 'Jumlah Penumpang'}, inplace=True)

    return passenger_count_hour

result = jumlah_penumpang_per_jam_weekday(df)
result
```

### 8.3 Filter DataFrame Berdasarkan Hari Kerja dan Jam Sibuk
Berdasarkan hasil analisis 8.1 dan 8.2, DataFrame akan difilter terlebih dahulu sebelum melakukan analisis lebih lanjut. Periode data yang dipilih adalah :
- Hari kerja = Senin - Jumat
- Jam sibuk = 05:00 - 09:00 dan 16:00 - 20:00

```python
# Mengonversi kolom tapInTime menjadi datetime
df['tapInTime'] = pd.to_datetime(df['tapInTime'])

# Memfilter untuk hari Senin sampai Jumat
df_weekday = df[df['tapInTime'].dt.day_name().isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])]

# Menentukan rentang waktu
pagi_start = pd.Timestamp('05:00:00').time()
pagi_end = pd.Timestamp('09:00:00').time()
sore_start = pd.Timestamp('16:00:00').time()
sore_end = pd.Timestamp('20:00:00').time()

# Memfilter DataFrame
df_filter = df_weekday[
    (df_weekday['tapInTime'].dt.time.between(pagi_start, pagi_end)) |
    (df_weekday['tapInTime'].dt.time.between(sore_start, sore_end))
]

# Menampilkan hasil filter
df_filter
```

### 8.4 Filter DataFrame berdasarkan `corridorID`
Langkah selanjutnya adalah memisahkan data antara moda TransJakarta dengan moda Royal Trans dan moda Microtrans. Dengan menggunakan filter berdasarkan payAmount untuk TransJakarta yaitu sebesar 3500 maka didapat data dengan moda TransJakarta.

```python
# Memfilter DataFrame untuk hanya menampilkan baris dengan payAmount == 3500
df_new = df_filter[df_filter['payAmount'] == 3500]

# Menampilkan DataFrame yang sudah difilter
df_new
```

### 8.5 Hitung dan Tambahkan Pemberhentian (`stop`)
Pemberhentian (`stop`) pada rute TransJakarta adalah titik atau lokasi di mana bus TransJakarta berhenti untuk menaikan dan menurunkan penumpang. Pemberhentian akan digunakan untuk perhitungan waktu ideal (`idealTime`)

```python
# Menambahkan kolom pemberhentian
df_new['stop'] = df_new['stopEndSeq'] - df_new['stopStartSeq']

df_new
```

### 8.6 Analisis Durasi Perjalanan (`travelTime`) setiap Koridor
Durasi perjalanan (`travelTime`) adalah waktu yang dihabiskan dalam perjalanan dari saat seorang penumpang masuk (tap in) hingga keluar (tap out). Durasi perjalanan didapat dari selisih antara `tapInTime` dengan `tapOutTime`. Satuan yang dipakai untuk durasi perjalanan adalah menit / min.

```python
# Menambahkan kolom 'durasiPerjalanan' dalam menit dan melakukan pembulatan ke atas
df_new['travelTime'] = np.ceil((df_new['tapOutTime'] - df_new['tapInTime']).dt.total_seconds() / 60).astype(int)

# Tampilkan DataFrame yang sudah ditambahkan
df_new
```

### 8.7 Kelompokan berdasarkan `corridorName` 
Mengelompokan durasi perjalanan berdasarkan `corridorName` dengan tujuan agar dapat mengetahui rata-rata durasi perjalanan untuk masing-masing koridor.

```python
# Menggunakan groupby
df_group = df_new.groupby('corridorName').agg({
    'corridorID': 'first', 
    'stop': 'median',  
    'travelTime': 'mean'         
}).reset_index()

# Membulatkan travelTime ke atas
df_group['travelTime'] = np.ceil(df_group['travelTime'])

df_group
```

### 8.9 Hitung dan Tambahkan Waktu Tempuh Ideal (`idealTime`)
Waktu tempuh ideal (`idealTime`) adalah perkiraan waktu yang diperlukan untuk menyelesaikan suatu perjalanan dalam kondisi optimal. Satuan yang digunakan pada waktu tempuh ideal adalah menit / min. Rumus menghitung waktu tempuh ideal sbb,

Waktu tempuh ideal = (Jarak (`distance`) / Kecepatan rata - rata) + total waktu berhenti 

Kecepatan rata-rata untuk TransJakarta adalah 21 km/jam.
Asumsi untuk waktu berhenti (termasuk waktu antri) adalah 6 menit / halte 
​


```python
# Menghitung waktu ideal
kecepatan = 21  # km/jam
df_analyze['idealTime'] = (df_analyze['distance'] / kecepatan) * 60  # dalam menit

# Menghitung waktu berhenti dan menambahkannya ke kolom idealTime
waktu_berhenti = 6  # dalam menit
df_analyze['idealTime'] += df_analyze['stop'] * waktu_berhenti  # menambahkan waktu berhenti

# Pembulatan ke atas
df_analyze['idealTime'] = np.ceil(df_analyze['idealTime'])

# Menampilkan DataFrame dengan kolom yang diperbarui
df_analyze

# Berdasarkan TomTom Traffic Index 2023, kecepatan rata-rata berkendara di Jakarta pada jam sibuk adalah 21 km/jam
# Asumsi untuk waktu berhenti (termasuk waktu antri) adalah 6 menit / halte 
```

# 9. Analisis Keterlambatan setiap Koridor
Keterlambatan (`delay`) dalam konteks layanan transportasi, khususnya TransJakarta, didefinisikan sebagai selisih antara durasi perjalanan aktual dan waktu ideal yang seharusnya dibutuhkan untuk menyelesaikan perjalanan. Untuk durasi perjalanan aktual yang kurang dari waktu ideal akan diisi dengan nilai 0. Analisis ini akan berfokus pada nilai keterlambatan lebih besar dari 0. Satuan yang digunakan dalam menit / min.

```python
# Menghitung keterlambatan
df_analyze['delay'] = df_analyze['travelTime'] - df_analyze['idealTime']

# Mengisi nilai negatif dengan nol
df_analyze['delay'] = df_analyze['delay'].clip(lower=0)

df_analyze
```

```python
# Menghapus nilai keterlambatan yang sama dengan atau kurang dari 0 dari DataFrame
df_barplot = df_analyze[df_analyze['delay'] > 0]

# Menghitung rata-rata keterlambatan berdasarkan corridorName
delay_data = df_barplot.groupby('corridorName')['delay'].mean().reset_index()

# Mengambil 5 keterlambatan tertinggi dan terendah
top_delay = delay_data.nlargest(5, 'delay')
bottom_delay = delay_data.nsmallest(5, 'delay')

# Menambahkan kolom untuk mengidentifikasi kategori
top_delay['Category'] = 'Keterlambatan Tertinggi'
bottom_delay['Category'] = 'Keterlambatan Terendah'

# Menggabungkan data
combined_delay = pd.concat([top_delay, bottom_delay])

# Mengatur ukuran figure
plt.figure(figsize=(10, 6))

# Membuat barplot horizontal dengan dua warna
palette = {'Keterlambatan Tertinggi': 'royalblue', 'Keterlambatan Terendah': 'lightcoral'}
barplot = sns.barplot(x='delay', y='corridorName', hue='Category', data=combined_delay, palette=palette)

# Mengatur label dan judul
plt.title('Perbandingan antara Keterlambatan Tertinggi dan Terendah')
plt.xlabel('Keterlambatan (menit)')
plt.ylabel('Nama Koridor')

# Menambahkan legend
plt.legend()

# Menampilkan nilai setiap batang
for p in barplot.patches:
    barplot.annotate(f'{p.get_width():.1f}', 
                     (p.get_width(), p.get_y() + p.get_height() / 2), 
                     ha='center', va='center', 
                     color='black', 
                     fontsize=10, 
                     xytext=(5, 0), 
                     textcoords='offset points')
    
# Menampilkan plot
plt.show()
```
